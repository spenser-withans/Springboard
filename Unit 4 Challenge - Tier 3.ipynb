**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(filtered_df):
    '''Find the price ratio between the years 1998 and 2018 for a dataframe filtered to a particular borough'''
#     Generating a dataframe that has of average price for each year
    price_avg_df = filtered_df.groupby('Year').agg({'Average_Price':np.mean})
#     Querying the price_avg_df for the price in 1998(starting_price)
    starting_price = price_avg_df.query('Year==1998').iloc[0,0]
#     Querying the price_avg_df for the price in 2018(ending_price)
    ending_price = price_avg_df.query('Year==2018').iloc[0,0]
#     Dividing and returning the ratio of the two prices.
    return ending_price/starting_price
```

```python
borough_list = list(df['London_Borough'].unique())
bor_price_dict = {}
for borough in borough_list:
    bor_price_dict[borough] =  create_price_ratio(df[df['London_Borough']== borough])
    
price_ratio = pd.DataFrame.from_dict(bor_price_dict,orient='index',columns = ['Price_Ratio_98_18'])
```

```python
df = df.join(price_ratio,on='London_Borough',how='left')
```

```python
df
```

```python
ratio_data = df.groupby('London_Borough').agg({'Price_Ratio_98_18':np.mean}).sort_values('Price_Ratio_98_18',ascending=False)

print(ratio_data)
print(ratio_data.describe())
ratio_data.hist(bins=10)
plt.show()
```

```python
def create_price_increase(filtered_df):
    '''Find the change in average house price from 1998 to 2018 for a particular london borough'''
#     Generating a dataframe that has of average price for each year
    price_avg_df = filtered_df.groupby('Year').agg({'Average_Price':np.mean})
#     Querying the price_avg_df for the price in 1998(starting_price)
    starting_price = price_avg_df.query('Year==1998').iloc[0,0]
#     Querying the price_avg_df for the price in 2018(ending_price)
    ending_price = price_avg_df.query('Year==2018').iloc[0,0]
#     Dividing and returning the ratio of the two prices.
    return ending_price-starting_price
```

```python
borough_list = list(df['London_Borough'].unique())
bor_price_dict = {}
for borough in borough_list:
    bor_price_dict[borough] =  create_price_increase(df[df['London_Borough']== borough])
    
price_increase = pd.DataFrame.from_dict(bor_price_dict,orient='index',columns = ['Price_Increase_98_18'])
```

```python
df = df.join(price_increase,on='London_Borough',how='left')
df
```

```python
price_increase = df.groupby('London_Borough').agg({'Price_Increase_98_18':np.mean}).sort_values('Price_Increase_98_18',ascending=False)
print(price_increase)
print(price_increase.describe())
price_increase.hist()
plt.show()
```
